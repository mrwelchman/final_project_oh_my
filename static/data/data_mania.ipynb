```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('nba_games.csv')
data.head()
```

```python
nba2020 = data.rename(columns={'Full Name (Data.Home Team)':'Home Team', 'Full Name (Data.Visitor Team)':'Visitor Team'})
nba2020.to_csv('nba2020.csv')
```

```python
nba2020['Point Difference'] = nba2020['Home Team Score']-nba2020['Visitor Team Score']
```

```python
nba2020['Home Win'] = np.where(nba2020['Point Difference'] > 0, 1, 0)
nba2020['Home Loss'] = np.where(nba2020['Point Difference'] < 0, 1, 0)
# nba2020.drop(columns=['home_win', 'home_loss'])
```

```python
nba2020.head()
```

```python
# # from sklearn.preprocessing import OneHotEncoder

# home_df=pd.get_dummies(nba2020['Home Team'], prefix='Home')
# visit_df=pd.get_dummies(nba2020['Visitor Team'], prefix='Visitor')

```

```python
df_visitor = pd.get_dummies(nba2020['Visitor Team'], dtype=np.int64)
df_home = pd.get_dummies(nba2020['Home Team'], dtype=np.int64)
```

```python
df_model = df_home.sub(df_visitor) 
df_model['Point Difference'] = nba2020['Point Difference']
```

```python
df_model.head()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
```

```python
df_train=df_model

lr = Ridge(alpha=0.001) 
X = df_train.drop(['Point Difference'], axis=1)
y=y = df_train['Point Difference']
X_train, X_test, y_train, y_test=train_test_split(X, y)

lr.fit(X_train, y_train)
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```
